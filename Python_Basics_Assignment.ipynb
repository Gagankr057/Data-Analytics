**Question 2 : Explain the difference between List and Tuple in Python.**

**Ans :**

- **List** : A List is a collection of items that is ordered and changeable (mutable).

Key Features:
- Allows duplicate values
- Elements can be modified
- Uses square brackets [ ]

- **Tuple** : A Tuple is a collection of items that is ordered but not changeable (immutable).

Key Features:
- Allows duplicate values
- Cannot modify elements after creation
- Uses parentheses ( )

**Question 3 : What is a function in Python? Why are functions useful?**

**Ans :** A function in Python is a block of code that performs a specific task and can be reused whenever needed. It is defined using the def keyword.

**Why are functions useful?**
- **Code Reusability** : You can use the same code multiple times without rewriting it.
- **Reduces Code Length**  : Makes programs shorter and cleaner.
- **Improves Readability** : Code becomes easier to understand and manage.
- **Modularity** : Breaks a big program into smaller parts.
