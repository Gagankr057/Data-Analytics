Q1. What is data cleaning, and why is it important in data analysis?
   
    - What are the potential consequences of analyzing unclean or messy data?   
    
    - Explain the common steps involved in cleaning and organizing data.

Ans :  
  -  Data cleaning is the process of detecting, correcting, or removing inaccurate, incomplete, inconsistent, or irrelevant data from a dataset.

  - It ensures that the data is accurate, consistent, and usable before analysis.

  - Data cleaning is important because:

     - **Better Accuracy**: Clean data gives correct insights and conclusions.

     - **Reliable Results**: Eliminates errors that could mislead your analysis.

     - **Better Model Performance**: Machine learning models perform better with clean data.

     - **Improved Decision-Making**: Businesses rely on accurate data for planning.

     - Well-cleaned data reduces issues during analysis or reporting.

  - Consequences of Analyzing Unclean or Messy Data - If you analyze messy data, you may face:

     - **Wrong conclusions**  → leading to bad decisions.

     - **Incorrect trends or patterns** → misleading insights.

     - **Poor model accuracy in machine learning.**

     - **Biased results due to missing or duplicated data.**

     - **Increased time and cost to fix wrong outputs later.**

     - **Loss of trust in data or analysis.**

  - Common Steps in Cleaning and Organizing Data

     - Remove duplicate Data.
     - Handle missing values → Delete blanks or fill using formulas
     - Fix data types → Convert text to numbers/dates.
     - Standardize text → Use TRIM, UPPER, PROPER, Find & Replace
     - Correct inconsistencies → Data Validation, consistent formats
     - Remove errors/outliers → Sort + Conditional Formatting
     - Split/merge columns → Text to Columns, CONCAT
     - Create calculated fields → Use formulas (Qty × Price)
     - Delete unnecessary rows/columns
     - Rename & organize columns → Clear headers, table format
     - Validate final data → Check totals, missing values

Q3. Explain the use of text functions such as TRIM , LEFT, RIGHT, MID, and CONCAT in data cleaning.

Ans : Use of Text Functions in Data Cleaning (Excel)

- TRIM() : Removes extra spaces from text.
     - Fix messy text with unwanted spaces
     - Clean names, product codes, IDs, etc.
     - Makes data consistent for comparison and lookup

- LEFT() : Extracts characters from the left side of a text.
     - Split codes or IDs
     - Extract first names (if fixed format)
     - Get prefixes (e.g., state codes like "DL12345" → "DL")

- RIGHT() : Extracts characters from the right side.
     - Use in data cleaning
     - Extract last digits of IDs
     - Get domain from email (only if fixed)
     - Example: RIGHT("A12345", 2) → "45"

4. MID() : Extracts characters from the middle of a text using start position and length.
     - Split text when structure is fixed
     - Extract part of a phone number, code, or ID
     - Useful when LEFT and RIGHT cannot handle it
     - Example : MID("ABC-123-XYZ", 5, 3) → "123"

5. CONCAT() : Joins two or more text strings together.
     - Combine first + last name
     - Create full address
     - Merge columns to create new IDs
     - Example: CONCAT("Gagan", " ", "Kumar") → "Gagan Kumar"

     - Restrict users from entering future dates

Q8. Understanding TODAY() and NOW()

 a. What is the difference between TODAY() and NOW() in Excel? Provide an example of when you would use each function.

 b. If cell A1 contains the date 2025-06-10, write a formula using TODAY() to determine how many days are left until that date

 c. Write an Excel formula using NOW() to display the current date and time in the format MM/DD/YYYY HH:MM AM/PM. ́

 d. If a cell contains =TODAY(), what will happen when the worksheet is reopened the next day? Explain

 e. You want to store a static date (today’s date) in a cell without it changing every day. What keyboard shortcut should you use?


Ans :

- Difference between TODAY() and NOW()
  - TODAY()
    - Returns only the current date (no time).
    - Example use: To calculate age, number of days remaining, or due dates
    - Example:  =TODAY() → 18-11-2025

  - NOW()
    - Returns current date AND current time.
    - Example use: To timestamp entries, track exact time of data entry, or build real-time dashboards.
    - Example :   =NOW() → 18-11-2025 11:47


- Formula to calculate days left until the date in A1 (2025-06-10)
  - If A1 contains 2025-06-10:   =A1 - TODAY()

- Use NOW() and format it as MM/DD/YYYY HH:MM AM/PM
  -  =TEXT(NOW(), "MM/DD/YYYY hh:mm AM/PM")   :  11/18/2025 11:54 AM

- If a cell contains =TODAY(), what will happen when the worksheet is reopened the next day
  - The value will automatically update to the new current date when the file is opened.TODAY() is a volatile function and always refreshes.

- Shortcut to insert a static date (non-changing date) :  ctrl + ;
