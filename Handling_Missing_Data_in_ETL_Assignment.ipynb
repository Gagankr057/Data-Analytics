# Handling Missing Data in ETL Assignment

**Q1. What are the most common reasons for missing data in ETL pipelines?**

**Ans :** The most common reasons for missing data in ETL pipelines are:

- **Incomplete data at source** – Some fields are optional or not captured by source systems.

- **Data entry errors** – Users may skip fields or enter invalid values.

- **System or extraction failures** – Network issues, crashes, or timeouts during data extraction.

- **Integration issues** – Mismatched schemas or joins across multiple data sources.

- **Data corruption or format mismatch** – Incorrect data types causing values to be dropped.

- **Privacy and compliance restrictions** – Sensitive fields intentionally left blank.

- **Transformation errors** – Logic mistakes during cleaning or mapping stages.

**Q2. Why is blindly deleting rows with missing values considered a bad practice in ETL?**

**Ans :** Blindly deleting rows with missing values is considered a bad practice in ETL because:

- It causes loss of valuable data, reducing dataset size.
- It can introduce bias, as missing data may follow a pattern.
- It hides important business insights related to why data is missing.
- It affects the accuracy of analytics and reports.
- It reduces the reliability of statistical analysis and ML models.
- It may remove valid records where only one non-critical field is missing.

**Q3. Explain the difference between:**

**Listwise deletion**

**Column deletion**

**Also mention one scenario where each is appropriate.**


**Ans :**

**Listwise Deletion** :
- Removes entire rows that contain one or more missing values.
- Ensures only complete records remain.

When appropriate : When the number of missing rows is very small and random, and deleting them does not impact analysis (e.g., 1–2 rows missing a critical field).

**Column Deletion**
- Removes an entire column that has many missing values.
- Used when the column adds little value.

When appropriate : When a column has excessive missing data (e.g., 70–80%) and is not important for reporting or modeling.

**Key difference :  Listwise deletion removes records (rows), while column deletion removes attributes (columns).**

**Q4. Why is median imputation preferred over mean imputation for skewed data such as income?**

**Ans :** Median imputation is preferred over mean imputation for skewed data like income because:

- Income data usually contains outliers (very high or very low values).
- The mean is highly affected by outliers, which can distort the true central value.
- The median is resistant to extreme values, making it more stable.
- Median better represents a typical income level in skewed distributions.
- It prevents inflated or unrealistic imputed values.

**Q5. What is forward fill and in what type of dataset is it most useful?**

**Ans :** Forward fill is a data imputation technique where a missing value is replaced with the last available non-missing value that appears before it in the dataset.

It is most useful in:

- Time-series data (sales, stock prices, sensor readings)
- Sequential or ordered datasets.
- Data where values change gradually over time.

**Q6. Why should flagging missing values be done before imputation in an ETL workflow?**

**Ans :** Flagging missing values should be done before imputation because:

- It preserves information about which values were originally missing
- Missingness itself may contain important business insights
- Helps track data quality issues during audits
- Improves machine learning model performance
- Allows analysis of patterns in missing data
- Prevents loss of transparency after imputation


**Q7. Consider a scenario where income is missing for many customers.How can this missingness itself provide business insights?**

**Ans :** Missing income data can itself reveal valuable business insights:

- Customers may be uncomfortable sharing financial details, indicating privacy concerns
- Certain regions or demographics may have higher non-disclosure rates
- Missing income may suggest self-employed, informal, or cash-based workers
- High-value or premium customers may intentionally avoid revealing income
- Indicates gaps in data collection processes or form design
- Helps businesses refine customer segmentation and targeting strategies
