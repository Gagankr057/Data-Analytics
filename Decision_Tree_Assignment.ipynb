**Q1. What is a Decision Tree? Explain its structure (Root, Nodes, Leaves) with a real-life example.**

**Ans :** A Decision Tree is a supervised machine learning algorithm used for classification and regression tasks. It works like a flowchart, where decisions are made by splitting the data based on feature values until a final prediction is reached.

Structure of a Decision Tree
**1. Root Node**
- The topmost node of the tree.
- Represents the entire dataset.
- It is the first feature used to split the data.

**2. Decision Nodes (Internal Nodes)**
- Intermediate nodes where further splitting occurs.
- Each node represents a condition or decision based on a feature.

**3. Leaf Nodes (Terminal Nodes)**
- The final nodes of the tree.
- Represent the predicted class or output value.
- No further splitting occurs.

**Real-Life Example: Bank Loan Approval**
- Suppose a bank wants to decide whether to approve a customer's loan application.

                 Income > 50,000?
                     /      \
                   Yes       No
                  /            \
       Credit Score > 700?     Reject
             /      \
           Yes       No
          /            \
      Approve         Reject


- Root Node: Income > 50,000?
- Decision Node: Credit Score > 700?
- Leaf Nodes: Approve Loan or Reject Loan

**Advantages of Decision Trees**
- Easy to understand and interpret.
- Handles both numerical and categorical data.
- Requires little data preprocessing.
- Can capture non-linear relationships.

**Disadvantages of Decision Trees**
* Can overfit the training data.
* Sensitive to small changes in data.
* May become complex when the tree grows very deep.


**Q2. Differentiate between Gini Impurity and Entropy.Which one is used by default in Scikit-learn and why?**

**Ans :** Both Gini Impurity and Entropy are measures used in Decision Trees to determine the best feature for splitting the data. They measure how mixed the classes are in a node.

**1. Gini Impurity**
- Measures the impurity or misclassification probability of a node.
- Indicates how often a randomly chosen element would be incorrectly classified.
* Value ranges from 0 (pure node) to 0.5 (maximum impurity in binary classification).
* Computationally simpler because it does not use logarithms.
* Faster to calculate than Entropy.
* Used in the CART (Classification and Regression Tree) algorithm.
* Default criterion in Scikit-learn (criterion='gini').
* Preferred when computational efficiency is important.
* Generally produces results similar to Entropy.

**2. Entropy**
- Measures the amount of uncertainty or randomness in a node.
- Based on Information Theory.
- Value ranges from 0 (pure node) to 1 (maximum uncertainty in binary classification).
* Uses logarithmic calculations.
* Computationally more expensive than Gini Impurity.
* Available in Scikit-learn using criterion='entropy'.
* Splits are chosen based on maximum Information Gain.
* Provides a stronger theoretical interpretation of information content.


**Gini Impurity is used by default because it is faster to compute and usually gives performance similar to Entropy.**
* Faster computation because it does not require logarithmic calculations.
* Poduces results very similar to Entropy in most practical cases.
* More computationally efficient for large datasets.

**Q3. What is Overfitting in Decision Trees?How can we detect it using training and testing accuracy?**

**Ans :** Overfitting occurs when a Decision Tree learns not only the important patterns in the training data but also the noise and random fluctuations. As a result, the model performs very well on the training data but poorly on unseen test data.

**Causes of Overfitting**
* Tree is too deep.
* Too many branches and nodes.
* Small number of samples in leaf nodes.
* Model memorizes the training data instead of learning general patterns.

How to Detect Overfitting?
* Compare Training Accuracy and Testing Accuracy:
* Training Accuracy: Performance on training data.
* Testing Accuracy: Performance on unseen test data.

**Case 1: Overfitting**
* Training Accuracy = 99%
* Testing Accuracy = 75%
* Large difference between training and testing accuracy indicates overfitting.

**Case 2: Good Fit**
* Training Accuracy = 82%
* Testing Accuracy = 80%
* Both accuracies are close, indicating good generalization.

**Case 3: Underfitting**
* Training Accuracy = 60%
* Testing Accuracy = 58%
* Both accuracies are low, indicating the model has not learned enough patterns.

**Q4. Explain Pruning in Decision Trees.What is the difference between Pre-pruning and Post-pruning?**

**Ans :** Pruning is the process of reducing the size of a Decision Tree by removing unnecessary branches or nodes. It helps prevent overfitting and improves the model's ability to generalize to new data.

**Why is Pruning Needed?**
* Reduces overfitting.
* Improves model performance on unseen data.
* Makes the tree simpler and easier to interpret.
* Reduces computational complexity.

Types of Pruning

**1. Pre-pruning (Early Stopping)**
* Pre-pruning stops the growth of the tree before it becomes too complex.

Common Parameters:

* max_depth
* min_samples_split
* min_samples_leaf
* max_leaf_nodes

**Advantages:**
* Faster training.
* Prevents overly complex trees.
* Reduces overfitting early.

**Disadvantages:**
* May stop tree growth too early.
* Can lead to underfitting.

**2. Post-pruning (Backward Pruning)**
* Post-pruning allows the tree to grow fully and then removes less important branches.

Process:
* Build the complete tree.
* Remove branches that contribute little to prediction accuracy.
* Keep the pruned tree if performance improves.

**Advantages:**
* Usually produces better generalization.
* Removes unnecessary branches after analyzing the full tree.

**Disadvantages:**

* More computationally expensive.
* Takes longer to train.

```python
# Q10. Model Building
# Train a Decision Tree Classifier

# Use:
  # criterion='gini'
  # max_depth=5

# Interpret:
# What does max_depth control?

df = pd.read_csv("/content/bank.csv")

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

for col in df.select_dtypes(include='object').columns:
    df[col] = label_encoder.fit_transform(df[col])

# Data spliting
X = df.drop('deposit', axis=1)
y = df['deposit']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42
)

# Model Training
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(
    criterion='gini',
    max_depth=5,
    random_state=42
)

dt_model.fit(X_train, y_train)
```

* A Decision Tree Classifier was trained using the training dataset.
* The Gini Impurity criterion was used to determine the best splits.
* The model learned patterns from customer information to predict whether a customer will subscribe to a term deposit.
* The trained model can now be used to make predictions on unseen data.

**What does max_depth control?**

* max_depth controls the maximum number of levels in the Decision Tree.
* It limits how deep the tree can grow.
* A larger depth can capture more complex patterns but may cause overfitting.
* A smaller depth reduces complexity and helps improve generalization.
* In this assignment, max_depth=5 helps balance model performance and overfitting.

```python
# Q11. Model Evaluation
# Perform:
# Accuracy score
# Confusion Matrix
# Classification Report

# Interpret (MUST) :
    # Is the model performing well?
    # Explain Precision & Recall
    # Any class imbalance issue?

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = dt_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

confusion = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(confusion)

report = classification_report(y_test, y_pred)
print("\nClassification Report:")
print(report)
```

**Is the model performing well?**
* Yes. The model achieves approximately 80.52% accuracy, indicating good predictive performance on unseen data.

**Precision :** Precison  measures how many of the predicted positive cases are actually positive.

Precision= (TP+FP)/TP
	​
* High precision means fewer false positives.

**Recall :** Recall measures how many actual positive cases are correctly identified.

Recall = (TP+FN)/TP
	​
* High recall means fewer false negatives.

```python
# Q12. Overfitting Check
# Calculate:
# Training accuracy
# Testing accuracy

# Interpret:
# Is the model overfitting or underfitting?

from sklearn.metrics import accuracy_score

y_train_pred = dt_model.predict(X_train)
y_test_pred = dt_model.predict(X_test)

train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

print("Training Accuracy:", train_accuracy * 100)
print("Testing Accuracy:", test_accuracy * 100)
```

* Training accuracy and testing accuracy are very close.
* The difference is approximately 1%, which is very small.
* The model performs consistently on both training and unseen test data.
* This indicates that the model has learned meaningful patterns rather than memorizing the training data.

**The model is neither overfitting nor underfitting.**

**Overfitting :** Occurs when training accuracy is much higher than testing accuracy.

**Underfitting :** Occurs when both training and testing accuracies are low.

**Current Model :** Training and testing accuracies are similar and reasonably high.

```python
# Q13. Pruning Experiment
# Train another model with:
# max_depth=3
# min_samples_split=20

# Interpret:
# Compare performance with previous model
# Which one is better and why?


from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

pruned_model = DecisionTreeClassifier(
    criterion='gini',
    max_depth=3,
    min_samples_split=20,
    random_state=42
)

pruned_model.fit(X_train, y_train)


# Trainig Accuracy

train_pred_pruned = pruned_model.predict(X_train)

train_acc_pruned = accuracy_score(
    y_train,
    train_pred_pruned
)

print("Training Accuracy:", train_acc_pruned)


# Testing Accuracy

test_pred_pruned = pruned_model.predict(X_test)

test_acc_pruned = accuracy_score(
    y_test,
    test_pred_pruned
)

print("Testing Accuracy:", test_acc_pruned)
```

**Perforamnce Comparison**
* The pruned model is simpler because the tree depth is reduced from 5 to 3.
* Training accuracy decreased from 81.50% to 77.78%.
* Testing accuracy decreased from 80.52% to 76.53%.
* The pruned model is less complex but loses some predictive power.
* Both models do not show significant overfitting.

**Which Model is Better and Why?**

- Model 1 (max_depth = 5) is better because:

* Higher training accuracy (81.50%).
* Higher testing accuracy (80.52%).
* Better overall predictive performance.
* No significant overfitting observed.
