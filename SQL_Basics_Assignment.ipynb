**Q9. What is the Difference Between a Primary Key and a Foreign Key?**

**Ans :**

**Primary Key**

- A Primary Key is a column (or a combination of columns) that uniquely identifies each record in a table.
It ensures that no two rows have the same value and that the value is never NULL.

- **Purpose**	: Uniquely identifies each record in a table
- **Uniqueness** : 	Must contain unique values
- **NULL values** : 	NULL values are not allowed
- Only one primary key per table
- **Data integrity**	: Ensures entity integrity
- Exists in the same table.

**Foreign Key**

- A Foreign Key is a column in one table that refers to the Primary Key of another table.
It is used to establish a relationship between two tables and maintain referential integrity.

-
- **Purpose**	: Links records between two tables
- **Uniqueness** : 	Can contain duplicate values
- **NULL values** : 	NULL values are allowed
- A table can have multiple foreign key.
- **Data integrity**	: Ensures refrential integrity
- Refers to another tables primary key


**Q10. What Are Constraints in SQL and Why Are They Used?**

**Ans :** Constraints in SQL are rules applied to table columns that restrict the type of data that can be stored in a database.
- They ensure data accuracy, consistency, and integrity.

- Constraints are used to:
    - Prevent invalid or incorrect data
    - Maintain data consistency
    - Enforce business rules
    - Protect data integrity
    - Improve database reliability

- Example of Types of Constraints in SQL

**1. PRIMARY KEY** : Ensures unique and non-NULL values

**2. FOREIGN KEY** : Links two tables and maintains relationships

**3. UNIQUE** : Ensures all values in a column are different

**4. NOT NULL** : Prevents NULL values
