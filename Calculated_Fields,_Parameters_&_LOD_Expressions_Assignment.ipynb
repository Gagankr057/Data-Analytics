**Question 1 : What is a Calculated Field and why is it important in business dashboards?**

**Ans :** A Calculated Field in Tableau is a new field created using formulas based on existing data.

**It is important because :**

- Helps derive new insights (e.g., Profit Ratio = Profit/Sales).
- Enables business logic (like classification, KPIs).
- Reduces dependency on raw dataset changes.
- Makes dashboards more dynamic and analytical.

**Question 2 :  Why can’t parameters change visuals directly in Tableau?**

**Ans :** Parameters in Tableau cannot directly change visuals because they are independent of the dataset.

- They simply act as input variables (containers for values) and are not connected to any fields or charts by default.

**Key Reason**
- Parameters do not filter or modify data on their own.
- They only store a value (like a number, string, or date).


**Question 3 : When should EXCLUDE LOD be used?**

**Ans :** EXCLUDE LOD is used when you want to remove a dimension from the level of detail in a calculation, even though that dimension is present in the view.

**Use EXCLUDE LOD when :**

- Compare detailed data vs overall summary
- Remove extra granularity (like Product) from calculation
- Show higher-level metrics inside a detailed view


**Question 4 : Difference between FIXED and INCLUDE LOD expressions?**

**Ans :** Both FIXED and INCLUDE are Level of Detail (LOD) expressions in Tableau, but they behave differently in terms of calculation level and filter interaction.

**FIXED LOD**
- Calculates values at a fixed level of detail defined in the formula
- Independent of the view (does not change with dimensions in view)
- Ignores regular filters (only context filters affect it)
- Used for benchmarking and consistent metrics
- Provides stable and constant results

**INCLUDE LOD**
- Adds extra dimensions to the current level of detail
- Depends on the view (changes with dimensions in view)
- Respects all filters
- Used for detailed and drill-down analysis
- Provides flexible and dynamic results
