**Question 1: What is a Null Hypothesis (H₀) and why is it important in hypothesis testing?**

**Ans :** The null hypothesis (H₀) is a statistical statement that assumes no effect, no difference, or no relationship exists in the population. It represents the status quo or the commonly accepted fact that researchers attempt to test.

- Importance of Null Hypothesis:
    - It provides a starting point for statistical testing.
    - All statistical tests are designed to assess whether there is enough evidence to reject H₀.
    - It helps in making objective decisions using data rather than personal judgment.
    - Without H₀, hypothesis testing would lack a clear decision framework.

- Example: If a company claims that the average complaint resolution time is 10 minutes, the null hypothesis is : H₀: μ = 10 minutes

**Question 2: What does the significance level (α) represent in hypothesis testing?**

**Ans :**
- The significance level (α) represents the maximum probability of committing a Type I error, i.e., rejecting a true null hypothesis.

- Key Points:
    - Common values of α are 0.05 (5%) and 0.01 (1%)
    - It acts as a decision threshold
    - If the p-value ≤ α, we reject H₀

Example: At α = 0.05, there is a 5% chance of incorrectly rejecting the null hypothesis, even when it is actually true.

**Question 3: Differentiate between Type I and Type II Errors**

**Ans :**
- Type I Error (α):
    - Occurs when the null hypothesis (H₀) is rejected even though it is true
    - Also called a false positive
    - Probability of Type I error is denoted by α (significance level)
    - Means concluding that an effect or difference exists when it actually does not

**Example:** Concluding a medicine is effective when it is not

- Type II Error (β):
    - Occurs when the null hypothesis (H₀) is not rejected even though it is false.
    - Also called a false negative
    - Probability of Type II error is denoted by β
    - Means failing to detect an actual effect or difference

**Example:** Concluding a medicine is ineffective when it actually works

**Question 4: Explain the difference between one-tailed and two-tailed tests with examples.**

**Ans :** In hypothesis testing, statistical tests are classified as one-tailed or two-tailed based on the direction of the alternative hypothesis.

- **One-Tailed Test**
   - A one-tailed test is used when the alternative hypothesis specifies a direction of effect, that is, it tests whether a parameter is greater than or less than a certain value.
   - Key Features:
       - Tests for an effect in only one direction
      - Entire significance level (α) is placed in one tail of the distribution
      - More powerful for detecting effects in the specified direction
Example : A company claims that the average delivery time is at most 10 minutes.

H₀: μ ≤ 10

H₁: μ > 10

- Here, we are only interested in knowing whether the delivery time exceeds 10 minutes, so a one-tailed test is appropriate.

- **Two-Tailed Test**
   - A two-tailed test is used when the alternative hypothesis does not specify a direction, and any deviation from the hypothesized value is considered important.
   - Key Features:
      - Tests for an effect in both directions
      - Significance level (α) is split equally between both tails
      - Used when changes on either side are meaningful

Example : A company claims that the average delivery time is exactly 10 minutes.

H₀: μ = 10

H₁: μ ≠ 10

Here, we test whether the delivery time is either greater than or less than 10 minutes, so a two-tailed test is used.

- Sample mean (x̄) = 12 minutes

Since |t| = 2 < 2.306, we fail to reject H₀.

**Question 6: When should you use a Z-test instead of a t-test?**

**Ans :**

- **Z-Test is used when:**
    - Sample size ≥ 30
    - Population standard deviation is known
    - Data is normally distributed

- **t-Test is used when:**
    - Sample size < 30
    - Population standard deviation is unknown
    - Sample standard deviation is used as an estimate
