**Q7: Variance for Decision-Making (Medium)
Two delivery companies track delivery delays (in minutes).**

**Company A: variance = 6**

**Company B: variance = 15**

**Which company is more consistent, and why?**

**Ans :** Variance measures how spread out the data is, **lower variance = less variation = more consistency.**
- Company A variance is 6 so it is more consistent.

- Company A’s delivery delays vary less from their average delay.

- Company B has a higher variance, meaning delivery times are more unpredictable.

- For customers and operations, lower variability usually means more reliable service.

- Therefore, Company A provides more consistent delivery performance.

**Q8: Standard Deviation in Context (Hard)
A finance student compares the daily price fluctuations of two cryptocurrencies.**

**Coin A: standard deviation = $30**

**Coin B: standard deviation = $120**
**Which coin is riskier to invest in, and why?**

**Ans :**
- Standard deviation measures how much prices fluctuate around the average.
- Higher standard deviation = higher volatility = higher risk.

- Comparison
   - Coin A: standard deviation = $30

   - Coin B: standard deviation = $120

- Conclusion : Coin B is riskier to invest in because :
   - Coin B’s price shows much larger daily fluctuations compared to Coin A.

   - This means prices can rise or fall sharply, increasing uncertainty.

   - While higher volatility can offer higher potential returns, it also brings a greater chance of loss.

- Therefore, due to its higher standard deviation, Coin B is the riskier investment.

**Q9: Combining Measures (Hard)
A family records their monthly electricity usage (in kWh): [400, 420, 390, 450, 410].Find the mean and standard deviation. What do these values together tell you about the family’s energy use
pattern?***

**Ans :**
- Mean = 2070/4 --> 414 kWh
- Standard Deviation :  Deviation from mean(414) : (-14,6, -24, 36, -4)
   - squares of deviations = (196,36,576,1296,16)
      - sum of squares = 2120
      - variance  = 2120/5 = 424
      - standard deviation  = square root of 424 = 20.6 kWh

**Interpretation (Mean + Standard Deviation together)**
- The mean (414 kWh) tells us the family’s typical monthly electricity usage.

- The standard deviation (≈20.6 kWh) shows that monthly usage usually varies by about ±21 kWh from the average.

- Since the standard deviation is small compared to the mean, the family’s electricity consumption is fairly consistent, with no extreme spikes or drops.

**Conclusion :** The family has a stable and predictable energy-use pattern, making it easier to plan budgets and estimate future electricity bills.
