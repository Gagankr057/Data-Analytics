Q1. Explain the difference between Relative , Absolute and Mixed Cell Referencing.
  - Relative Cell Reference(A1)
    - changes when you copy the formula to another cell.
    - Excel automatically adjusts the row and column.

  - Absolute Cell Reference ("dollar sign"A"dollar sign"1)
    - Does not change when you copy the formula
    - Both column and row is fixed using $ (dollar is palced before coulumn and row when we enter the formula).

  - Mixed Cell Reference ($A1) or (A"dollar sign"1)
    - Either the row or column is fixed.
    - $A1 - Column A is fixed
    - A$1 - row 1 is fixed
