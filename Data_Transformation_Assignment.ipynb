```python
# Q1. Handle Missing Values (Basic Cleaning)
# Fill missing values using the following:
# - Age → Median
# - Salary → Mean
# - Hours_Worked_Per_Week → Median
# - Performance_Score → Mean

# Display the dataset after handling missing values.


# Ans :

df = pd.read_csv("/content/employee_productivity_dataset.csv")

missing_age = df['Age'].median()
df['Age'] = df['Age'].fillna(missing_age)

missing_salary = df['Salary'].mean()
df['Salary'] = df['Salary'].fillna(missing_salary)


missing_hours = df['Hours_Worked_Per_Week'].median()
df['Hours_Worked_Per_Week'] = df['Hours_Worked_Per_Week'].fillna(missing_hours)

missing_score = df['Performance_Score'].mean()
df['Performance_Score'] = df['Performance_Score'].fillna(missing_score)


print(df.head(10))
```

```python
# Q2. Label Encoding
# Convert these columns into numeric values using Label Encoding:
# - Gender
# - Department
# Show the updated columns.


# Ans :

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df['Gender'] = label_encoder.fit_transform(df['Gender'])

df['Department'] = label_encoder.fit_transform(df['Department'])

print(df[['Gender', 'Department']].head(10))
```

```python
# Q3. One-Hot Encoding
# Apply One-Hot Encoding on:
# - Work_Mode
# - Location
# Display the dataset and check how many new columns are created.

# Ans :

df = pd.read_csv("/content/employee_productivity_dataset.csv")
df = pd.get_dummies(
    df,
    columns=['Work_Mode', 'Location']
)


print(df.head())

print("\nTotal Number of Columns:")
print(df.shape[1])
```

```python
# Q4. Normalization (Min-Max Scaling)

# Normalize the following columns:

# - Salary
# - Hours_Worked_Per_Week

# Use MinMaxScaler and display results.


# Ans :

df = pd.read_csv("/content/employee_productivity_dataset.csv")

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df[['Salary', 'Hours_Worked_Per_Week']] = scaler.fit_transform(
    df[['Salary', 'Hours_Worked_Per_Week']]
)

print(df[['Salary', 'Hours_Worked_Per_Week']].head(10))
```

```python
# Q5. Standardization (Scaling)

# Apply StandardScaler on:

# - Age
# - Projects_Completed

# Display the transformed values.

# Ans :

df = pd.read_csv("/content/employee_productivity_dataset.csv")

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[['Age', 'Projects_Completed']] = scaler.fit_transform(
    df[['Age', 'Projects_Completed']]
)


print(df[['Age', 'Projects_Completed']].head(10))
```

```python
# Q7. Build Preprocessing Pipeline

# Create a pipeline that:

# - Applies encoding to categorical columns
# - Applies scaling to numerical columns

# Use:
# ColumnTransformer
# Pipeline


# Ans :

import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from sklearn.preprocessing import (
    OneHotEncoder,
    StandardScaler
)

df = pd.read_csv("/content/employee_productivity_dataset.csv")

categorical_columns = [
    'Gender',
    'Department',
    'Work_Mode',
    'Location'
]

numerical_columns = [
    'Age',
    'Salary',
    'Hours_Worked_Per_Week',
    'Performance_Score',
    'Projects_Completed'
]

numerical_pipeline = Pipeline([
    ('scaler', StandardScaler())
])

categorical_pipeline = Pipeline([
    ('encoder', OneHotEncoder())
])

preprocessor = ColumnTransformer([
    ('num', numerical_pipeline, numerical_columns),
    ('cat', categorical_pipeline, categorical_columns)
])

print("Pipeline Created Successfully")
print(preprocessor)
```

```python
# Q8. Apply Pipeline
# Apply the pipeline on the dataset.

# Display:
# Transformed dataset
# Shape of final dataset

# Ans :


import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from sklearn.preprocessing import (
    OneHotEncoder,
    StandardScaler
)

df = pd.read_csv("/content/employee_productivity_dataset.csv")

categorical_columns = [
    'Gender',
    'Department',
    'Work_Mode',
    'Location'
]

numerical_columns = [
    'Age',
    'Salary',
    'Hours_Worked_Per_Week',
    'Performance_Score',
    'Projects_Completed'
]

numerical_pipeline = Pipeline([
    ('scaler', StandardScaler())
])

categorical_pipeline = Pipeline([
    ('encoder', OneHotEncoder())
])

preprocessor = ColumnTransformer([
    ('num', numerical_pipeline, numerical_columns),
    ('cat', categorical_pipeline, categorical_columns)
])

transformed_data = preprocessor.fit_transform(df)

print(transformed_data)

print("\nShape of Final Dataset:")
print(transformed_data.shape)
```

**Q9.Why is scaling important in Python?**

**Ans :** Scaling is important in Machine Learning because different features may have different value ranges. Large-value features can dominate small-value features and affect model performance.

For example:
- Salary : values in lakhs
- Age : values around 20–50

Without scaling, algorithms may give more importance to Salary than Age.

**Benefits of Scaling**
- Improves model accuracy
- Speeds up training process
- Prevents large-value features from dominating
- Makes gradient descent converge faster

**Common scaling methods:**
- MinMaxScaler
- StandardScaler

**Q10. Why do we convert categorical data into numerical form?**

**Ans :** Machine Learning algorithms cannot understand text or categorical values directly.They only work with numerical data.

Ex:- Gender : Male, Female

- A machine learning model cannot process "Male" or "Female" as text.

- So we convert them into numbers using encoding techniques.

**Common Encoding Methods**

**1. Label Encoding**
- Converts categories into numbers.

 Example:
   - Male : 1
   - Female : 0

**2. One-Hot Encoding**

- Creates separate columns for each category

Example:
- Work_Mode_Remote
- Work_Mode_Hybrid
- Work_Mode_Office


**Benefits**
- Makes data understandable for machine learning models
- Improves model performance
- Helps algorithms process categorical features correctly
