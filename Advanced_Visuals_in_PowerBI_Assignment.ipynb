**Q.1 What is the difference between a Basic Filter, Advanced Filter, and a Slicer in Power BI?**

**Ans :**

**1. Basic Filter**

- Used to filter data by selecting specific values from a list.
- It allows simple filtering using checkboxes.
- Example: Selecting only 2023 from the Year column.

**2. Advanced Filter**

- Used for complex filtering conditions such as:
   - greater than
   - less than
   - contains
   - between

Example: Sales greater than ₹50,000.

**3. Slicer**

- A visual filtering tool placed directly on the report page.

- Users can interactively filter data by clicking options.

Example: A slicer showing Region (India, USA, Europe).

**Q4. What is a Visual-Level Filter, and how does it differ from Page-Level and Report-Level Filters?**

**Ans :**

**Visual-Level Filter**

- A Visual-Level Filter is a filter that applies only to a single visual (chart, table, or graph) in a Power BI report.
- It affects only that specific visual
- Other visuals on the same page remain unchanged.

Example : If you apply a filter Region = India to a Bar Chart, only that bar chart will show sales from India, while other visuals may still show global sales.

**Page-Level Filter**

- A Page-Level Filter applies to all visuals on a single report page.
- Every chart, table, or graph on that page will follow the filter.

Example : If you apply Year = 2023 as a page-level filter, every visual on that page will show only 2023 sales data.

**Report-Level Filter**

- A Report-Level Filter applies to all pages in the entire Power BI report.
- It affects every visual across every page.

Example : If you apply Country = India as a report-level filter, all visuals on every page will display sales data only from India.

**Q6. What is the role of Drill-Down in Power BI? How does it improve data exploration?**

**Ans :** Drill-Down in Power BI is a feature that allows users to navigate from higher-level summary data to more detailed levels of data within a hierarchy.

It helps users analyze data step by step by moving through levels such as:

Year → Quarter → Month → Day

or

Country → Region → City

**Role of Drill-Down**

**1. Detailed Analysis** : Users can explore deeper levels of data from a summarized view.

**2. Better Data Understanding** : It helps identify patterns, trends, and problem areas.

**3. Interactive Exploration** : Users can click on visuals to move between hierarchy levels without creating multiple charts.

**4. Efficient Decision Making** : Managers can quickly find which category, region, or time period affects performance.

**Q10. Explain the difference between Drill-Down and Drill-Through in Power BI using a real-world example.**

**Ans :**

**Drill-Down**
- Drill-Down allows users to move from summary data to more detailed levels within the same visual using a hierarchy.

Example hierarchy : Year → Quarter → Month → Day

**Real-World Example** : A sales manager views total yearly sales.
Using Drill-Down, they click on 2023 to see quarterly sales, then monthly sales, helping them identify which month had the highest or lowest sales.

**Drill-Through**
- Drill-Through allows users to navigate to another report page that contains more detailed information about a selected data point.

**Real-World Example** : A manager sees sales by product on the main report page.When they right-click on "Laptop", they select Drill Through to open another page showing region-wise sales of the Laptop product.

**Q13. What is the “Keep All Filters” option in Drill-Through? How does it affect user interaction**

**Ans :**

**Keep All Filters in Drill-Through**

- “Keep All Filters” is an option in Power BI that ensures all filters applied on the main report page are passed to the drill-through page.
- This means when a user navigates to the drill-through page, the same filtering context remains active.

**How It Affects User Interaction**

1. Maintains Context
     - All selected filters such as Region, Product Category, or Year remain applied when moving to the drill-through page.

2. More Accurate Analysis
     - Users see detailed data that matches their current selection.

3. Better User Experience
     - Users do not need to reapply filters on the drill-through page.
