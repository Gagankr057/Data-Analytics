```python
# Q1. Create a countplot to show the number of passengers by gender.

# Ans :

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/Titanic.csv")

plt.figure(figsize=(6,4))
sns.countplot(x='Sex', data=df, palette='Set2')

plt.title("Number of Passengers by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
#Q2. Create a bar chart to show the number of passengers in each passenger class using different colors.

# Ans :

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("/content/Titanic.csv")

class_counts = df['Pclass'].value_counts().sort_index()


plt.figure(figsize=(6,4))
plt.bar(class_counts.index,
        class_counts.values,
        color=['skyblue', 'lightgreen', 'red'])


plt.title("Number of Passengers in Each Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")


plt.xticks([1, 2, 3])

plt.show()
```

```python
# Q3. Create a histogram to show the distribution of passenger age.

# Ans :

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/Titanic.csv")

plt.figure(figsize=(8,5))
sns.histplot(df['Age'].dropna(),
             bins=30,
             kde=True,
             color='skyblue')

plt.title("Distribution of Passenger Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
# Q4. Create a boxplot to detect outliers in the fare column.

# Ans :


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/Titanic.csv")

plt.figure(figsize=(7,5))
sns.boxplot(y='Fare', data=df, color='lightblue')

plt.title("Boxplot of Passenger Fare")
plt.ylabel("Fare")

plt.show()
```

```python
# Q5. Create a scatterplot to

#  a. show the relationship between age and fare.

#  b. analyze the relationship between family size (SibSp + Parch) and fare.


# Ans :

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/Titanic.csv")

plt.figure(figsize=(8,5))
sns.scatterplot(x='Age', y='Fare', data=df, color='blue')

plt.title("Relationship Between Age and Fare")
plt.xlabel("Age")
plt.ylabel("Fare")

plt.show()
```

```python
# Q6. Create a countplot to show survival count by gender

# Ans :

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/Titanic.csv")

plt.figure(figsize=(7,5))
sns.countplot(x='Sex', hue='Survived', data=df, palette='Set1')

plt.title("Survival Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")

plt.legend(title='Survived', labels=['No', 'Yes'])

plt.show()
```

```python
# Q7. Create a bar chart to show survival by passenger class.

# Ans :

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/Titanic.csv")

survival_class = df.groupby('Pclass')['Survived'].sum()

plt.figure(figsize=(7,5))
plt.bar(survival_class.index,
        survival_class.values,
        color=['gold', 'silver', 'brown'])

plt.title("Survival Count by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Survivors")

plt.xticks([1, 2, 3])

plt.show()
```

```python
# Q8. Create a heatmap to show correlation between numerical variables.

# Ans :

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/Titanic.csv")

numerical_data = df.select_dtypes(include=['int64', 'float64'])

corr_matrix = numerical_data.corr()

plt.figure(figsize=(10,7))
sns.heatmap(corr_matrix,
            annot=True,
            cmap='coolwarm',
            fmt='.2f')

plt.title("Correlation Heatmap of Numerical Variables")

plt.show()
```

```python
# Q9. Identify which passenger class paid the highest fare on average.

# Ans :

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/Titanic.csv")

avg_fare = df.groupby('Pclass')['Fare'].mean()

print(avg_fare)

plt.figure(figsize=(7,5))
plt.bar(avg_fare.index,
        avg_fare.values,
        color=['gold', 'silver', 'brown'])

plt.title("Average Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Fare")

plt.xticks([1, 2, 3])

plt.show()
```
