# Power BI Basics Assignment

**Question 1: What is Power BI and why is it used in businesses?**

**Ans :** Power BI is a business intelligence and data visualization tool developed by Microsoft. It helps organizations collect data from multiple sources, analyze it, and present insights through interactive reports and dashboards.

- Why businesses use Power BI:
     - To make data-driven decisions
     - To visualize complex data easily
     - To track KPIs and performance
     - To combine data from different sources
     - To share insights securely across teams

**Question 2: Name and explain the three main components of Power BI**


**Ans :**

**1. Power BI Desktop**
- A Windows application used to connect data, clean data, create data models, and build reports.

**2. Power BI Service (PowerBI.com)**
- A cloud-based platform used to publish, share, and collaborate on reports and dashboards.

**3. Power BI Mobile Apps**
- Mobile applications for Android and iOS that allow users to view dashboards and reports anytime, anywhere.

**Question 3: Explain the Power BI workflow**


**Ans :**  The Power BI workflow includes the following steps:

- **Data Connection** – Connect to data sources like Excel, SQL Server, CSV, etc.

- **Data Transformation** – Clean and transform data using Power Query.

- **Data Modeling** – Create relationships and measures using DAX.

- **Report Creation** – Build visualizations and reports.

- **Publishing** – Publish reports to Power BI Service.

- **Sharing & Collaboration** – Share dashboards with stakeholders.

**Question 4: List any four data cleaning tasks that can be performed in Power Query**

**Ans :**
- Removing duplicate rows
- Handling missing or null values
- Changing data types (text, number, date)
- Splitting or merging columns

**Question 5: Write step-by-step instructions to load the Real_Estate dataset**

**Ans :**

- Open Power BI Desktop
- Click on Get Data
- Select the data source (Excel / CSV / Text – depending on dataset)
- Click Browse and select the Real_Estate dataset
- Click Open
- Preview the data in the Navigator window
- Click Load
- Dataset is now loaded into Power BI

**Question 6: Define Data View, Report View, and Model View**


**Ans :**

**Data View**
- Displays the dataset in table format
- Used to inspect data and create calculated columns

**Report View**
- Used to create visualizations like charts, tables, and dashboards
- Main area for report building

**Model View**
- Used to manage relationships between tables
- Helps in building a proper data model

**Question 7: Discuss the different data sources that Power BI supports**

**Ans :** Power BI supports multiple data sources, including :

- **Files** : Excel, CSV, XML, PDF
- **Databases** : SQL Server, MySQL, Oracle, PostgreSQL
- **Online Services** : SharePoint, Google Analytics, Azure
- **Cloud Platforms** : Azure SQL Database, Azure Blob Storage
- **Web Sources** : APIs and web pages
