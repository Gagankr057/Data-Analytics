**Q1. What is a Common Table Expression (CTE), and how does it improve SQL query readability?**

**Ans :**

- A CTE (Common Table Expression) is a temporary named result set defined using the '**WITH**' clause. It exists only during the execution of a query.
- it improves readability by :
    - Breaks complex queries into logical steps
    - Avoids repeated subqueries
    - Makes debugging and maintenance easier
    - Looks similar to writing modular code

- Example :

WITH HighPriceProducts AS (
  SELECT ProductID, ProductName, Price
  FROM Products
  WHERE Price > 2000
)
SELECT * FROM HighPriceProducts;


**Q2. Why are some views updatable while others are read-only? Explain with an example.**

**Ans :** A view is a virtual table created using a SELECT statement.

- **Updatable Views** : A view is updatable when changes made through the view can be directly reflected in the base table.
- Updatable views allow INSERT, UPDATE, and DELETE operations when they are simple and map directly to a single table.
- A view is updatable if:
     - It is based on only one table
     - It does not use: GROUP BY, DISTINCT,Aggregate functions (SUM, AVG, COUNT, etc.), JOIN, Calculated/derived columns
     - It includes the primary key of the base table

Example :
CREATE VIEW vw_ProductsSimple AS
SELECT ProductID, ProductName, Price
FROM Products;

Updating through the view :
UPDATE vw_ProductsSimple
SET Price = 1300
WHERE ProductID = 1;

- **Read-Only Views** : A view becomes read-only when SQL cannot uniquely identify which row(s) in the base table should be updated.
- Read-only views are used only for reporting and analysis because they involve aggregation, grouping, or joins.
- A view is read-only if it contains:
     - GROUP BY
     - Aggregate functions (SUM, AVG, COUNT)
     - JOIN of multiple tables
     - DISTINCT
     - Calculated columns

Example :
CREATE VIEW vw_CategorySummary AS
SELECT Category, COUNT(ProductID) AS TotalProducts
FROM Products
GROUP BY Category;

**Q3. What advantages do stored procedures offer compared to writing raw SQL queries repeatedly?**

**Ans :** A stored procedure is a precompiled set of SQL statements stored in the database and executed as a single unit. Stored procedures provide several advantages over writing raw SQL queries repeatedly.

- Advantages of Stored Procedures

**1. Better Performance**
- Stored procedures are compiled once and reused
- Reduces query parsing and execution time
- Faster than executing raw SQL repeatedly

**2. Reusability**
- Write once, use many times
- Same procedure can be called by multiple applications or users

**3. Improved Security**
- Users can be given permission to execute procedures without direct access to tables.
- Helps prevent SQL injection
- Sensitive logic remains hidden

**4. Reduced Network Traffic**
- Only the procedure name and parameters are sent over the network
- Avoids sending long SQL queries multiple times

**5. Easier Maintenance**
- Business logic is centralized in the database
- Any change needs to be made only in the procedure, not in every application

**6. Support for Business Logic**
- Allows use of:Conditional statements (IF, CASE)
- Loops
- Variables
- Makes complex operations easier to manage

**Stored procedures improve performance, security, reusability, and maintainability, making them more efficient and reliable than writing raw SQL queries repeatedly.**


**Q4. What is the purpose of triggers in a database? Mention one use case where a trigger is essential.**

**Ans :**
- A trigger is a special database object that automatically executes in response to specific events on a table or view, such as INSERT, UPDATE, or DELETE.
- Triggers are used to enforce rules, maintain data integrity, and automate actions without requiring manual execution.

**Purpose of Triggers**

- Automatic Execution
     - Runs automatically when a specified event occurs
     - No need for user or application intervention

- Maintain Data Integrity
    - Ensures consistency of data across tables
    - Prevents invalid data operations

- Enforce Business Rules
    - Validates or restricts operations based on conditions
    - Example: Prevent deletion of important records

- Audit and Logging
     - Keeps track of changes (who changed what and when)

- Synchronization of Tables
    - Automatically updates related tables when data changes

**Use Case of a Trigger**

Use Case: Archiving Deleted Records
- When a product is deleted from the Products table, a trigger automatically stores the deleted record in an archive table for auditing and recovery purposes.

**Q5. Explain the need for data modelling and normalization when designing a database.**

**Ans :** Data Modelling and Normalization are essential steps in database design to ensure that data is organized, accurate, efficient, and scalable.

**Need for Data Modelling**

- Data modelling is the process of defining:
   - Tables
   - Columns
   - Relationships between tables

- Why data modelling is needed
    - Clear Structure : Helps understand how data is stored and connected
    - Defines Relationships : Establishes primary keys and foreign keys
    - Reduces Redundancy : Avoids storing the same data in multiple places
    - Improves Data Integrity : Ensures valid and consistent data
    - Better Scalability : Makes future changes easier

Example: Separating Products and Sales into different tables instead of storing everything in one table.

**Need for Normalization** : Normalization is the process of organizing data to minimize duplication and dependency.

- Why normalization is needed
    - Eliminates Data Redundancy : Prevents duplicate data storage
    - Avoids Data Anomalies : Prevents update, insert, and delete anomalies
    - Improves Data Consistency : Same data remains uniform across the database
    - Saves Storage Space : Efficient use of disk space
    - Enhances Data Integrity : Logical and accurate data structure


**Data modelling provides a logical blueprint of the database, while normalization ensures efficient, consistent, and error-free data storage. Together, they form the foundation of a reliable database design.**
