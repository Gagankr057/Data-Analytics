**Question 4: The population of a school is divided into 60% boys and 40% girls. If you want equal representation
of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling?
Why?**


**Ans :** We should use **Stratified Sampling** because :
 - The population is not evenly distributed (60% boys, 40% girls).
 - Stratified Sampling divides the population into strata (boys and girls) and then selects equal numbers from each group, ensuring balanced representation of both genders.
 - We should **not** use **Simple Random Sampling**,because the sample may end up with more boys than girls, leading to unequal representation.

**Question 10: Differentiate between discrete and continuous random variables with examples.**

**Ans :**
- **Discrete Random Variable :**   
    - Takes countable values
    - Values are finite or countably infinite
    - Obtained by counting
    - Values are usually whole numbers
    - Probability of a specific value can be non-zero
    - Uses Probability Mass Function (PMF)
      - Examples:
         - Number of students in a class
         - Number of heads in coin tosses
         - Number of defective items

- **Continuous Random Variable**
  - Takes uncountable values
  - Values lie in a continuous range
  - Obtained by measurement
  - Values can be fractional or decimal
  - Probability of a specific value is always zero
  - Uses Probability Density Function (PDF)

     - Examples:
        - Height of a person
        - Weight of an object
        - Time taken to complete a task
        - One-line Memory Trick
