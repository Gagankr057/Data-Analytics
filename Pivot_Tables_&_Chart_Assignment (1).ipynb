**Q1. What is the purpose of a PivotTable in Excel?**

**Ans:**
- A PivotTable in Excel is used to quickly summarize, analyze, organize, and compare large datasets without changing the original data. It helps you extract meaningful insights by automatically grouping, sorting, filtering, counting, and calculating data.

**Q3. Apply Conditional Formatting to highlight sales greater than 400.**

**Ans:**
- Select the Sales column
- Click and drag to select all cells under the Sales heading (not the header).
- Go to the menu:
    - Home → Conditional Formatting → Highlight Cells Rules → Greater Than…
    - A dialog box will appear.
    - In the box, type : **400**
- Choose a formatting style
- Click OK.

**Q4. Explain the difference between Sorting and Filtering with an example from this dataset**

**Ans:**

**Sorting**  : Sorting means rearranging the entire dataset based on one or more columns — either in ascending (A–Z / smallest to largest) or descending (Z–A / largest to smallest) order.



**Filtering** : Filtering means showing only the rows that meet a condition and hiding the rest, while the data stays in the same order.
