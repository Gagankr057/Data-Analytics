**Q8. What is the difference between a Worksheet and a Dashboard in Tableau?
Why is a dashboard more useful for decision-makers?**

**Ans :**

**Worksheet :**
- A Worksheet is a single view or visualization.
- It is used to create individual charts like bar chart, line chart, map, etc.
- Example: Sales by Category chart

**Dashboard :**
- A Dashboard is a combination of multiple worksheets.
- It displays multiple visualizations in one screen.
- Example: Sales, Profit, and Region charts together.

**Dashboard is more useful for decision-makers because :**

- Provides a complete overview of business data.
- Combines multiple insights in one place.
- Helps in quick and informed decision-making.
- Supports interactive analysis (filters, drill-downs)

**Q9.  What is Tableau Prep, and why is it important in real-world data projects?**

**Then explain how you would:**

- Remove null values
- Remove duplicate records.
- Change incorrect data types (Answer in steps, no tool access required.)

**Ans :**
Tableau Prep is a data preparation tool used to:

- Clean data
- Combine data
- Transform data before analysis

**Why is it important in real-world projects?**

- Ensures accurate and reliable data
- Removes errors and inconsistencies
- Improves quality of analysis
- Saves time during visualization

 Data Cleaning Steps

**1. Remove Null Values**

- Identify fields with null values
- Apply Filter
- Exclude Null values

This ensures clean and usable data

**2. Remove Duplicate Records**

- Identify duplicate rows
- Use Remove Duplicates option
OR
- Apply aggregation/grouping to eliminate duplicates

**3. Change Incorrect Data Types**

- Identify incorrect data types (e.g., text instead of number)
- Click on the field
- Change data type:
   - String → Number
   - String → Date

This ensures correct calculations and analysis.
