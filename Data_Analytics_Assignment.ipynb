Q3. Write the end-to-end analytics workflow in the correct order and briefly explain each step.
   
  - These are the steps in end-to-end data analytics workflow

      -> Problem Definition : Fisrt step is to define the problem and clearly understand the what problem needs solving.
      
      -> Data collection : Next step is to gather the data related to problem from different sources such as Excel sheets , Sql dtabases.

      -> Data Prepration : Next step is to clean and transform the raw data and add missing values and removing errors and transform raw data into meaningful insights.

      -> Data Visulazation : Next step is presenting the meaningful insights into chart and graphs as it more visually appealing and easy to understand.

      -> Decision Analysis : Last step is data analysis and making decison according to the useful insights gain in above steps and make necessary changes.



Q4. What is Prompt Engineering, and explain any two types of prompts used in Generative AI?
   
  -  Prompt Engineering is the process of designing,refining,and structuring input prompts given to a generative AI model like ChatGPT to get the most accurate or creative output.
  
 -  Two common types of Prompts in Generative AI:-
      
      -> Zero-shot Promt :- this model is given no examples , only the instruction, the AI uses its pre-trained knowledge to generate the answer.

      -> Few-shot Prompt :- this prompt is given few examples before the main query, it helps the model understand the format,tone or style expected.

Q6. Explain any three AI-powered ETL (Extract, Transform, Load) tools and how they are used in analytics.
  
  - ETL stands for Extract, Transform, and Load - it's the process of:
    1. Extracting data from multiple sources
    2. Transforming it into a usable format
    3. Loading it into a data warehouse or anlytics platform

 - Three AI powered ETL tools are :
    - Talend Data Fabrics :- Talend is a popular cloud based ETL and Data integration platform that uses AI and Machine learning to enhance data quality , automate cleansing and optimize workflow

    - Informatica Intelligent Data Management Cloud(IDMC) :- Informatica IDMC is a AI-driven ETL and Data integration platform powered by its AI engine, CLAIRE. It automates repetitive ETL tasks and helps analysts quickly prepare data for insights

    - Apache Hop :- Apache Hop is an open source ETL and data orchestration platform that integrates with AI/ML models to automate transformation steps and data validation.

Q8.Explain the evolution of analytics, highlighting the different stages from Descriptive to Generative AI.

   - Analytics has evolved through several stages , each adding more intelligence, automation and decision making capability.

   1. Descriptive Analytics :- What happened? - Past focused
      - Descriptive analytics summarizes historical data to explain what has already happened.
         - Sales report
         - monthly revenue charts

   2. Diagnostic Analytics :- Why did it happen?
      - Diagnostic analytics digs deeper into data to explain why something happened.
         - Why did sales drop last quarter?
  
   3. Predictive Analytics :- What will happen?
      - Predictive analytics uses statical models and amchine learning to forecast future outcomes.
         - Predicting customer loss
         - Forecasting demand or stock prices
         - Predicting equipment failure

   4. Prescriptive analytics :- What should we do?
      - prescriptive analyitcs recommends optimal actions based on predictions and constraints.
         - Best Pricing strategy
         - Optiimal delivery routes
         - Inventory optimization

   5. Cognitive Analytics - How can machines think like humans?
      - Cognitive analytics uses AI,NLP and deep learnig to mimic human thinking and understanding.
         - understand languages : Chatbots
         - image recognition

   6. Generative AI - Create new content or insights
      - Generative AI creates new content - Text,images,code,insights-based on traning data.
          - Generate texts, reports, and summaries.
          - create images,design and videos.
          - Generate code,test cases.
