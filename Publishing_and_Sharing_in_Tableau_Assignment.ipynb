# Publishing and Sharing in Tableau Assignment

**Question 1. What is the key difference between publishing a dashboard on Tableau Public versus Tableau Server/Cloud in terms of accessibility and data security?**

**Ans :**

**Tableau Public :**
- Dashboards are accessible to anyone on the internet.
- All data is publicly visible.
- No data security or privacy control.
- Open access, no security.

**Tableau Server / Tableau Cloud :**
- Dashboards are accessible only to authorized users.
- Provides strong data security (login, permissions, roles).
- Suitable for confidential and business data.
- Controlled access, high security.

**Question 2. When publishing a Tableau dashboard, what happens to a live connection compared to an extract?How does this affect data refresh and performance?**

**Ans :**

**Live Connection :**
- Connects directly to the original data source (database).
- Data is always up-to-date (real-time).
- No need to refresh manually.
- Performance can be slower, especially with large or complex queries.


**Extract :**
- Stores a snapshot of data inside Tableau.
- Requires scheduled/manual refresh to update data.
- Provides much faster performance.
- Works even when the original database is offline.

**Question 3. In Tableau Server or Cloud (conceptually), which feature allows administrators to control which users can view, interact with, or edit a published dashboard?**

**Ans :** The feature used is **Permissions** (User Roles & Access Control)

**It allows administrators to :**
- Control who can view dashboards.
- Decide who can interact (filter, explore).
- Allow or restrict editing and downloading.
- Assign roles like Viewer, Explorer, Creator.

**Question 4. If you want users to view a dashboard but not download the underlying data, which permission setting would you adjust in Tableau Server?**

**Ans :** We need to adjust the “Download Data” permission in Tableau Server/Cloud, Specifically : Disable “Download Full Data” / “Download Underlying Data”.

**Question 5. Name two methods by which a Tableau dashboard can be embedded into a web page or website.**

**Ans :** There are two common methods to embed a Tableau dashboard into a web page:

1. **Using Embed Code (iframe):**
   Tableau provides an HTML iframe embed code that can be copied and pasted directly into a website’s HTML. This is the simplest method and does not require programming knowledge.

2. **Using JavaScript API (Tableau Embedding API):**
   This method allows more advanced control over the dashboard. Developers can add interactivity such as applying filters, responding to user actions, and customizing the dashboard behavior using JavaScript.



**Question 6. List two best practices you should follow when preparing a dashboard for public sharing to ensure data security and privacy.**

**Ans :** When preparing a Tableau dashboard for public sharing, it is important to follow these best practices:

1. **Remove Sensitive or Confidential Data:**
- Ensure that no personal, financial, or confidential business information is included in the dashboard before publishing it publicly.

2. **Use Data Anonymization or Aggregation:**

- Modify the data by removing identifiable details (such as names, phone numbers, or IDs) and present it in aggregated form (e.g., totals, averages) to protect individual privacy.

**Question 7. After publishing a dashboard, users report a “Data not found” error. What is the most common cause related to data source settings?**

**Ans :** The most common cause of a “Data not found” error is a broken or incorrect data source connection.

**This usually happens when:**

- The file path or database location has changed.
- The data source is not accessible from Tableau Server/Cloud.
- Credentials were not embedded or are incorrect.

**Question 8. When embedding a Tableau dashboard into a secure internal portal, what two additional security measures should be implemented to ensure only authorized users can access the embedded content?**

**Ans :** When embedding a Tableau dashboard into a secure internal portal, the following security measures should be implemented:

1. **User Authentication (Login / Single Sign-On - SSO):**
- Ensure that only authenticated users can access the dashboard by requiring login credentials or integrating Single Sign-On (SSO).
2. **Row-Level Security (RLS) or Permissions:**
- Apply permissions or row-level security so that users can only view the data they are authorized to see.
