```python
# Q1. What is the average, median, and standard deviation of house prices?

# Ans :

df = pd.read_csv("/content/House Price India.csv")

average_price = df['Price'].mean()
median_price = df['Price'].median()
std_price = df['Price'].std()

print("Average House Price:", average_price)
print("Median House Price:", median_price)
print("Standard Deviation of House Price:", std_price)
```

```python
# Q2. Which number of bedrooms is most common?

# Ans :

df = pd.read_csv("/content/House Price India.csv")

most_common_bedroom = df['number of bedrooms'].mode()[0]

print("Most Common Number of Bedrooms:", most_common_bedroom)
```

```python
# Q3. Check if the price data is skewed.

# Ans :

df = pd.read_csv("/content/House Price India.csv")

price_skewness = df['Price'].skew()

print("Skewness of Price Data:", price_skewness)
```

- The price data is highly positively skewed because the skewness value is greater than 1.

**This means :**
- Most houses are priced in the lower-to-middle price range.
- A small number of very expensive luxury houses create a long right tail in the distribution.
- The dataset is not normally distributed.

```python
# Q4. What is the average price for each number of bedrooms?

# Ans :

df = pd.read_csv("/content/House Price India.csv")

avg_bedroom_price = df.groupby('number of bedrooms')['Price'].mean()

print("Average Price for Each Number of Bedrooms:", avg_bedroom_price.round(2))
```

- Average price generally increases with the number of bedrooms.
- 7 and 8-bedroom houses have the highest average prices, indicating luxury properties.
- Prices for 9, 10, and 11-bedroom houses do not consistently increase, likely because there are very few such houses in the dataset.
- The 33-bedroom house is highly unusual and likely:
    - a data entry error,
    - incorrect dataset value , or an extreme outlier.

```python
#Q5.  What is the relationship between living area and price? (Use : Area without Basement).

# Ans :

df = pd.read_csv("/content/House Price India.csv")

relationship = df['Area of the house(excluding basement)'].corr(df['Price'])

print("Correlation between Living Area and Price:", relationship)
```

- The correlation value 0.6152 indicates a moderately strong positive relationship between living area and house price.

- This means that as the living area increases, house prices generally tend to increase as well.

- Larger houses usually have higher prices because they provide more usable space and better facilities.
- Since the value is not extremely close to 1, other factors such as location, waterfront, condition, and number of bedrooms also influence house prices.

```python
# Q6. Identify any anomalies where houses have high prices but low areas.

# Ans :

df = pd.read_csv("/content/House Price India.csv")

anomalies = df[
    (df['Price'] > df['Price'].quantile(0.85)) &
    (df['Area of the house(excluding basement)'] < df['Area of the house(excluding basement)'].quantile(0.20))
]

print("Anomalies with High Prices and Low Areas : ","\n ", anomalies)
```

- These houses are considered anomalies or outliers.

**Possible reasons:**
- Premium location
- Waterfront/property view
- Luxury interiors
- Historical or commercial value
- Data entry errors

```python
# Q7. Compare average price based on number of floors and houses with or without waterfront.

# Ans :

df = pd.read_csv("/content/House Price India.csv")

comparison = df.groupby(['number of floors', 'waterfront present'])['Price'].mean()

print(comparison.round(2))
```

- Houses with waterfront views usually have higher average prices.
- Multi-floor houses may also cost more because they generally provide larger living space.
- Waterfront properties are considered premium real estate and attract higher market value.
- Comparing these groups helps identify which property features influence pricing the most.

```python
# Q8. Identify the minimum and maximum house price. What does this indicate?

# Ans :

df = pd.read_csv("/content/House Price India.csv")

min_price = df['Price'].min()
max_price = df['Price'].max()

print("Minimum House Price:", min_price)
print("Maximum House Price:", max_price)
```

- The minimum price represents the most affordable property in the dataset (78000).

- The maximum price represents the most expensive or luxury property(7700000).

- A large gap between minimum and maximum prices indicates high variation in the housing market.

**This variation may occur due to:**
- Location differences
- House size
- Waterfront availability
- Luxury features
- Property condition

```python
# Q9. Which location (zipcode/area) has the highest average price?

#  Ans :

df = pd.read_csv("/content/House Price India.csv")

avg_price_location = df.groupby('Postal Code')['Price'].mean()

highest_price_location = avg_price_location.idxmax()
highest_avg_price = avg_price_location.max()

print("Location with Highest Average Price:", highest_price_location)
print("Highest Average Price:", round(highest_avg_price, 2))
```
