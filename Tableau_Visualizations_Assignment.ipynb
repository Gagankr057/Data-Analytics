**Q9. Chart Selection Justification**

**Explain why different charts were used for different business questions in this analysis.**

**Ans :** Different charts were used in the analysis based on the type of data and the business objective. Each visualization was selected to clearly represent insights and support decision-making.

**1. Bar Chart (Category Sales & Regional Performance)**
- Used to compare sales across categories and regions
- Helps in identifying highest and lowest performers
- Provides clear and accurate comparison between discrete values

**2. Line Chart (Sales Trend & Growth Analysis)**
- Used for time-based analysis
- Shows sales trends and patterns over months
- Running total helps evaluate overall business growth

**3. Scatter Plot / Alternative (Profitability Analysis)**
- Used to analyze relationship between Sales and Profit
- Helps identify:
    - Loss-making orders
    - High sales but low profit
- (Alternatively, bar chart can be used for better clarity in case of dense data)

**4. Stacked / Side-by-Side Bar Chart (Category vs Region)**
- Used to compare multiple dimensions together
- Helps identify regional performance for each category
- Useful for spotting strong and weak market areas

**5. Bubble Chart (Product Prioritization)**
- Used for multi-dimensional analysis
- Represents:
   - Sales (X-axis)
  - Profit (Y-axis)
  - Quantity (Size)
- Helps identify high-impact products

**6. Dashboard**
- Combines multiple visualizations into one interactive view
- Includes filters like Region and Category
- Enables quick and efficient decision-making

**Q10. Business Insight Summary : Summarize two key insights from your dashboard and explain how they can help management make better decisions.**


**Ans :**

**Insight 1: High Sales Does Not Always Mean High Profit**
- Analysis shows that some products or orders generate high sales but low peofit.
- This indicates issues like:
   - Heavy discounts
   - High operational or product costs
   
**Business Impact :** Can reduce overall profitability despite strong sales numbers.

**Recommendation:**
- Optimize pricing strategy
- Reduce excessive discounts
- Focus on high-profit products instead of just high sales


**Insight 2: Uneven Regional Performance**
- Certain regions contribute significantly more to total sales than others.
- Some regions show low sales and lower profitability.

**Business Impact :** Indicates untapped market potential or weak regional strategy.

**Recommendation :**
- Invest more in high-performing regions
- Improve marketing, logistics, and product availability in low-performing regions.
- Use region-specific strategies.
