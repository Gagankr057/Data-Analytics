# Data Quality and Validation in ETL Assignment

**Question 1 : Define Data Quality in the context of ETL pipelines. Why is it more than just data cleaning?**

**Ans :** Data Quality in ETL refers to the degree to which data is accurate, complete, consistent, valid, timely, and reliable as it moves through Extract, Transform, and Load stages.

It is more than just data cleaning because:
- Cleaning fixes visible issues (NULLs, duplicates, formats)
- Data quality ensures business correctness
- It validates rules, relationships, and constraints
- Poor-quality data can be “clean” but still wrong

**Question 2 : Explain why poor data quality leads to misleading dashboards and incorrect decisions.**

**Ans :** Poor data quality directly impacts the reliability of dashboards and reports. When data contains duplicates, missing values, incorrect formats, or invalid relationships, the calculated metrics become inaccurate.

- Duplicate records inflate totals such as sales, revenue, or customer count.
- Missing or NULL values distort averages, trends, and KPIs.
- Incorrect or inconsistent data leads to wrong comparisons across time, regions, or products.
- Invalid master–transaction relationships result in incomplete or broken reports.

**Question 3 : What is duplicate data? Explain three causes in ETL pipelines.**

**Ans :** Duplicate data refers to the presence of multiple records representing the same real-world entity or transaction in a dataset. Duplicates cause inaccurate reporting, inflated metrics, and unreliable analytics.

Three common causes of duplicate data in ETL pipelines:

**1. Multiple data sources**:
The same data may arrive from different source systems without proper reconciliation, creating repeated records.

**2. Improper incremental load logic**:
When ETL jobs reprocess historical data instead of only new or changed records, the same rows get loaded multiple times.

**3. Missing or incorrect business keys** :
If unique identifiers (business keys) are not properly defined or enforced, ETL processes cannot detect and eliminate duplicates.

**Question 4 : Differentiate between exact, partial, and fuzzy duplicates.**

**Ans :**

**Exact Duplicates**
- All columns or business key values are exactly the same
- Easy to detect using SQL GROUP BY
- Example: Same Customer_ID, Product_ID, Txn_Date, and Txn_Amount

**Partial Duplicates**
- Some fields match, while others differ
- Occur due to inconsistent updates or missing attributes
- Example: Same customer and product but different city or quantity

**Fuzzy Duplicates**
- Records are similar but not exactly identical
- Caused by spelling mistakes or formatting differences
- Example: Rahul Mehta vs Rahul M.

**Question 5 : Why should data validation be performed during transformation rather than after loading?**

**Ans :**  
- Prevents incorrect or invalid data from entering the target system.

- Reduces the need for costly data corrections and reprocessing.

- Improves overall ETL performance and efficiency.

- Allows early detection and handling of data errors.

- Ensures only business-rule-compliant data is loaded.

- Protects dashboards and reports from bad data.

**Question 6 : Explain how business rules help in validating data accuracy. Give an example.**

**Ans :**
- Business rules define what data values are valid and acceptable for the organization.
- They ensure data follows real-world and domain logic
- Help detect incorrect, illogical, or inconsistent data
- Prevent invalid data from being loaded into target systems
- Improve trust in reports and analytics

Example:

- Quantity must be greater than 0
- Transaction amount cannot be NULL
- Transaction date cannot be NULL or in the future

Records violating these rules are treated as invalid data and handled during ETL.
