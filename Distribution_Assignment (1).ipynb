**Q4. What is the concept of standardization (z-score), and why is it important in data analysis? Explain the
formula and how standardization transforms a dataset.**


**Ans :**
- Standardization is the process of converting data values from their original scale into Z-scores, which represent how many standard deviations a value is from the mean of the dataset.

- Z Score Formula
   -  **Z = (Orignal Data value - Mean)/ standard deviation**

- When a dataset is standardized:
   - The mean becomes 0
   - The standard deviation becomes 1
   - The shape of the distribution remains the same
   - Only the scale changes, not the relative spacing of values


- Standardization Is Important in Data Analysis because :

   1. Makes data comparable
       - Allows comparison between datasets with different units or scales(e.g., marks vs. height, income vs. age).

   2. Helps interpret position of values
       - Shows whether a value is above or below average and by how much.

   3. Required for probability calculations
       - Enables use of the standard normal (Z) table to find probabilities.

   4. Essential in statistical methods
       - Used in hypothesis testing, confidence intervals, and outlier detection.

**Q5. What is Kurtosis and their type?**

**Ans :**
- Kurtosis is a statistical measure that describes the shape of a frequency distribution, focusing on:
    - Peakedness (how sharp or flat the peak is)
    - Tail thickness (how heavy or light the tails are)

- It tells us how data values are concentrated around the mean and how extreme values (outliers) behave compared to a normal distribution.

- There are three main types of kurtosis:

1. **Mesokurtic**
- Distribution similar to the normal distribution
- Moderate peak and tails

**Kurtosis value:**
- Kurtosis = 3
- Excess kurtosis = 0

     - Example: Normal distribution

2. **Leptokurtic**
- Sharp, high peak
- Heavy (thick) tails
- More extreme values (outliers)

**Kurtosis value:**
- Kurtosis > 3
- Excess kurtosis > 0
    - Example: Income distribution, stock market returns

3. **Platykurtic**
- Flat or low peak
- Thin tails
- Fewer extreme values

**Kurtosis value:**
- Kurtosis < 3
- Excess kurtosis < 0
    - Example: Uniform distribution


**Q6.Explain why the uniform distribution is a good model for the outcome of rolling a fair die.**

**Ans :**
- A uniform distribution is a good model for rolling a fair die because all six outcomes (1 to 6) have an equal chance of occurring. Each number has the same probability, **1/6**, and no outcome is favored. Therefore, the results of a fair die follow a uniform distribution.
- In a discrete uniform distribution, all outcomes in the sample space are equally likely. This matches exactly how a fair die behaves.

**Q8.How does log transformation help in stabilizing variance and making data more normally distributed?**

**Ans :**
- Log transformation is a mathematical technique where we replace each data value X with log(X). It is commonly used when data is skewed or shows unequal variance.

**1. Stabilizing Variance**
- In many datasets, variance increases with the mean (common in income, sales, population data).
- Log transformation compresses large values more than small values.
- This reduces the spread of high values, making variance more constant across the data.
    - Result: Heteroscedastic data â†’ more homoscedastic

**2. Making Data More Normally Distributed**
- Right-skewed data (long tail to the right) becomes more symmetric after log transformation.
- Extreme values (outliers) are pulled closer to the center.
- The distribution often becomes closer to a normal (bell-shaped) curve.

**Example :**
- Original values: 10,20,50,200,1000

After log transformation : 1,1.3,1.7,2.3,3

ðŸ‘‰ Large gaps shrink, improving symmetry.

**This Is Important because :**
- Many statistical methods (regression, correlation, hypothesis tests) assume normality and constant variance.
- Log transformation helps data meet these assumptions.
- It improves model accuracy and interpretation.
- Simple Assignment Answer
- Log transformation reduces skewness by compressing large values and stabilizes variance by reducing the effect of extreme observations, often making the data more normally distributed.
