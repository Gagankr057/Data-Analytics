**Q1. Define Covariance and explain how it differs from Correlation in terms of scale and interpretation.**

**Ans :**

**Covariance** : Covariance is a statistical measure that indicates the direction of the linear relationship between two variables. It shows whether variables increase or decrease together.
  - Covariance depends on units.
  - It's range is from (-infinity to +infinity).
  - It is difficult to interpret.
  - Covariance is not comparable because it depends on units and scale.

**Correlation** : Correlation is a standardized form of covariance that measures both direction and strength of the linear relationship.
  - Correlation is unit free.
  - It ranges from (-1 to +1).
  - Interpretation is easy.
  - Correlation is comparable because it is standardized and unit-free.

**Correlation is preferred because it provides a clear, standardized interpretation.**

**Q2. What does a positive, negative, and zero covariance indicate about the relationship between two variables?**

**Ans :** Covariance indicates the direction of the linear relationship between two variables.

**1. Positive Covariance**
- Indicates that both variables move in the same direction
- When one variable increases, the other also increases
- When one variable decreases, the other also decreases
    - Example : Height and weight, income and expenditure

**2. Negative Covariance**
- Indicates that variables move in opposite directions
- When one variable increases, the other decreases, and vice versa
     - Example : Price and demand, speed and travel time

**3. Zero Covariance**
- Indicates no linear relationship between the variables
- Changes in one variable do not systematically affect the other
     - Example : Shoe size and exam marks

**Q3. Discuss the limitations of covariance as a measure of relationship between two variables. Why is correlation preferred in many cases?**

**Ans :**
- **Limitations of Covariance**

**1. Depends on units of measurement** : The value of covariance changes if the units of variables change (e.g., cm to m, ₹ to $).

**2. Difficult to interpret magnitude** : A higher or lower covariance value does not clearly indicate how strong the relationship is.

**3. Not comparable across datasets** : Since covariance is not standardized, relationships between different variables cannot be compared.

**4. Only shows direction, not strength** : Covariance only tells whether variables move together or oppositely, not how strongly.

- **Why Correlation is Preferred**

**1. Unit-free and standardized** : Correlation is independent of measurement units.

**2. Easy to interpret** : Values lie between –1 and +1, making interpretation straightforward.

**3. Comparable across datasets** : Strength of relationships between different variable pairs can be compared.

**4. Measures both direction and strength** : Correlation clearly indicates how strongly and in which direction variables are related.

Covariance has limitations due to unit dependency and lack of standardization, whereas correlation overcomes these issues by providing a standardized, unit-free, and easily interpretable measure of relationship.

**Q4. Explain the difference between Pearson’s correlation coefficient and Spearman’s rank correlation coefficient. When would you prefer to use Spearman’s correlation?**

**Ans :**

**1. Pearson’s Correlation Coefficient** : Pearson’s correlation coefficient measures the strength and direction of a linear relationship between two continuous numerical variables.

- Key Features:
   - Measures linear relationship
   - Uses actual data values
   - Assumes data is normally distributed
   - Highly sensitive to outliers

**2. Spearman’s Rank Correlation Coefficient** : Spearman’s rank correlation coefficient measures the strength and direction of a monotonic relationship using ranked data.

- Key Features:
    - Measures monotonic (not necessarily linear) relationship
    - Uses ranks instead of actual values
    - Does not require normal distribution
    - Less affected by outliers

**Spearman’s correlation is preferred when:**
- Data is in rank form
- Relationship is non-linear but monotonic
- Dataset contains outliers
- Data does not follow normal distribution

**Q5. If the correlation coefficient between two variables X and Y is 0.85, interpret this value in context. Can you infer causation from this value? Why or why not?**

**Ans :** A correlation coefficient of 0.85 indicates a strong positive linear relationship between variables X and Y.

**This means:**
- As X increases, Y also tends to increase
- As X decreases, Y also tends to decrease
- The relationship between X and Y is strong and in the same direction.
- No, causation cannot be inferred from this value.


**Reason**

- **Correlation does not imply causation** : Correlation only measures association, not cause-and-effect.

- **Third variable problem** : A third factor may influence both X and Y.

- **Reverse causality** : It is unclear whether X causes Y or Y causes X.

- **Coincidental relationship** : High correlation can occur by chance.
