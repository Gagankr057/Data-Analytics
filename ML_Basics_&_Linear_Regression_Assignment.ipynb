**Q1. What is Machine Learning? Explain the main types of Machine Learning with examples.**

**Ans :** Machine Learning (ML) is a branch of Artificial Intelligence that enables computers to learn patterns from data and make predictions or decisions without being explicitly programmed.

**Types of Machine Learning**

**1. Supervised Learning :** The model learns from labeled data (input and output are known).

Examples:

Examples:
- Customer segmentation

Examples:
- Self-driving cars

**Q2. What is Supervised Learning? Explain the difference between Regression and Classification.**

**Ans :**

**Supervised Learning** : Supervised Learning is a type of Machine Learning in which a model is trained using labeled data. In labeled data, both the input features and the correct output values are provided. The model learns the relationship between inputs and outputs during training and then uses this knowledge to make predictions on new, unseen data. Examples of supervised learning include predicting house prices, detecting spam emails, and forecasting sales.

**Difference Between Regression and Classification**

**Regression**
- Regression is used to predict continuous numerical values.
- The output is a number.
- Used when the target variable is continuous.
- Examples: House price prediction, stock price prediction, temperature forecasting.
- Common algorithms: Linear Regression, Ridge Regression, Lasso Regression.
- Evaluation metrics: MAE, MSE, RMSE, R² Score.
- Example output: House Price = ₹45,00,000.
- Predict Numerical Values


**Classification**
* Classification is used to predict categorical values or classes.
* The output is a label/category.
* Used when the target variable is categorical.
* Examples: Spam detection, disease diagnosis, sentiment analysis.
* Common algorithms: Logistic Regression, Decision Tree, Random Forest, Naive Bayes.
* Evaluation metrics: Accuracy, Precision, Recall, F1-Score.
* Example output: Email = Spam or Not Spam.
* Predict Category or class.


**Q3. What is Linear Regression? Write the Linear Regression equation and explain intercept and coefficient.**

**Ans :**
**Linear Regression**

Linear Regression is a supervised machine learning algorithm used to predict a continuous dependent variable based on one or more independent variables. It establishes a linear relationship between the input features and the target variable by fitting the best possible straight line through the data points.


**Linear Regression Equation**   y=β0+β1x1+β2x2+----+βnxn
	​

Where:

* y = Predicted value (dependent variable)
* β₀ = Intercept
* β₁, β₂, ..., βₙ = Coefficients of independent variables
* x₁, x₂, ..., xₙ = Independent variables (features)

**Intercept (β₀)** : The intercept is the predicted value of the dependent variable when all independent variables are equal to zero. It represents the point where the regression line crosses the y-axis.

**Coefficient (β₁, β₂, ..., βₙ)** : A coefficient represents the change in the dependent variable for a one-unit increase in the corresponding independent variable, while keeping all other variables constant.

Example:
If the coefficient of the number of rooms (RM) is 4.44, it means that for every additional room, the house price is expected to increase by approximately 4.44 units, assuming all other factors remain unchanged.

**Q4. Explain the following evaluation metrics used in Linear Regression:**

* **SSE**
* **MSE**
* **RMSE**
* **R² Score**
* **Adjusted R²**

**Ans :** Evaluation metrics are used to measure how well a Linear Regression model performs by comparing actual values with predicted values.

**1. SSE (Sum of Squared Errors)** : SSE is the sum of the squared differences between actual values and predicted values. It measures the total error made by the model.
- A lower SSE indicates a better model fit.

**2. MSE (Mean Squared Error)** : MSE is the average of the squared errors. It tells us how far the predictions are from the actual values on average.
- Lower MSE values indicate better model performance.

- Advantage: Penalizes large errors more heavily because errors are squared.

**3. RMSE (Root Mean Squared Error)** : RMSE is the square root of MSE. It measures prediction error in the same units as the target variable.
- A lower RMSE indicates more accurate predictions.

- Advantage: Easier to interpret because it is expressed in the original units of the target variable.

**4. R² Score (Coefficient of Determination)** : R² Score measures the proportion of variance in the dependent variable that is explained by the independent variables.

* R² = 1 → Perfect prediction
* R² = 0 → Model explains no variance
* Higher R² → Better model performance

**5. Adjusted R²** : Adjusted R² is a modified version of R² that considers the number of independent variables in the model. It prevents overestimating model performance when unnecessary features are added.

* Higher Adjusted R² indicates a better model.
* It increases only when a new feature genuinely improves the model.

**Q5. Explain the Machine Learning workflow steps from data collection to model evaluation.**

**Ans :** The Machine Learning workflow is a systematic process used to build and evaluate a machine learning model. It involves several steps, from collecting data to assessing the model's performance.

**1. Data Collection**

* The first step is to gather relevant data from various sources such as databases, surveys, websites, sensors, or business records. The quality of data directly affects the model's performance.

**2. Data Preprocessing (Data Cleaning)**

* Raw data often contains missing values, duplicate records, and inconsistencies. Data preprocessing involves cleaning the data, handling missing values, removing duplicates, and converting data into a suitable format for analysis.

**3. Exploratory Data Analysis (EDA)**

* EDA is performed to understand the dataset and discover patterns, trends, and relationships between variables. Techniques such as summary statistics, pairplots, histograms, and heatmaps are commonly used.

**4. Feature Selection and Feature Engineering**

* Relevant features are selected to improve model performance. New features may also be created from existing variables to provide more meaningful information to the model.

**5. Splitting the Dataset**

* The dataset is divided into training and testing sets. Typically, 80% of the data is used for training and 20% for testing. This helps evaluate how well the model performs on unseen data.

**6. Model Selection**

* A suitable machine learning algorithm is chosen based on the problem type. For example, Linear Regression is used for regression problems, while Logistic Regression is used for classification problems.

**7. Model Training**

* The selected algorithm is trained using the training dataset. During training, the model learns patterns and relationships between input features and the target variable.

**8. Model Prediction**

* After training, the model makes predictions on the test dataset. These predictions are compared with actual values to assess performance.

**9. Model Evaluation** : The model's performance is measured using evaluation metrics. For Linear Regression, common metrics include:

* MAE (Mean Absolute Error)
* MSE (Mean Squared Error)
* RMSE (Root Mean Squared Error)
* R² Score
* Adjusted R²

**10. Model Improvement and Deployment**

* If the model's performance is not satisfactory, improvements such as feature engineering, hyperparameter tuning, or selecting a different algorithm may be applied. Once the model performs well, it can be deployed for real-world use.

```python
# Q10. Train a Linear Regression model and perform the following:
# Make predictions
# Calculate MAE
# Calculate MSE
# Calculate RMSE
# Calculate R² Score
# Calculate Adjusted R²
# Write the Linear Regression equation
# Interpret the model coefficients

# Ans :

# 1: Train the Linear Regression Model

from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)

# 2: Make Predictions

y_pred = model.predict(X_test)

print("Predicted Values:", y_pred[:5])

# 3: Create Evaluation Metrics

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Adjusted R²
n = len(y_test)
p = X_test.shape[1]

adj_r2 = 1 - ((1-r2)*(n-1)/(n-p-1))

print("\nMAE:", mae)
print("\nMSE:", mse)
print("\nRMSE:", rmse)
print("\nR² Score:", r2)
print("\nAdjusted R²:", adj_r2)

# 4. linear regression Equation

print("\nIntercept:", model.intercept_)
print("\nCoefficients:", model.coef_)
# #MEDV = β₀ + β₁(CRIM) + β₂(ZN) + β₃(INDUS)
#       + β₄(CHAS) + β₅(NOX) + β₆(RM)
#       + β₇(AGE) + β₈(DIS) + β₉(RAD)
#       + β₁₀(TAX) + β₁₁(PTRATIO)
#       + β₁₂(B) + β₁₃(LSTAT)


```
