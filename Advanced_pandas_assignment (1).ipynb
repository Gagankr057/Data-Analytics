```python
# Q2 Sales Filtering Analysis
# Filter the dataset to display:
# - Orders where Sales > 3500
# - Orders where Region = "East" and Sales > 2500
# Explain what insights you observe.


# Ans :
df = pd.read_csv('/Advanced_pandas_sales_dataset.csv')

high_sales = df[df['Sales']>3500]
print("\nHigh Sales", "\n",high_sales)

East_sales = df[(df['Region'] == 'East') & (df['Sales'] > 2500)]
print("\nEast Sales : ","\n" ,East_sales)
```

**Ans 2 :**  Insights  :
- High-value transactions are concentrated in electronics products, especially laptops.
- The East region shows strong performance, making it a potential focus area for business expansion.
- Businesses should focus on premium products and high-performing regions to maximize revenue.

```python
# Q3 Customer Purchase Analysis
# Identify the top 5 customers based on total sales.
# Hint: Use GroupBy with aggregation.

# Ans :

df = pd.read_csv('/Advanced_pandas_sales_dataset.csv')

top_customers = df.groupby('Customer')['Sales'].sum().sort_values(ascending=False).head(5)

print(top_customers)
```

```python
# Q4 Regional Sales Performance

# Calculate the total sales generated by each region.

# Then identify: Which region generated the highest revenue?


#Ans :

df = pd.read_csv('/Advanced_pandas_sales_dataset.csv')

region_sales = df.groupby('Region')['Sales'].sum()
print(region_sales)

highest_revenue_region = region_sales.idxmax()
print("\nHighest Revenue Region : ",highest_revenue_region)
```

```python
# Q5 Product Performance Analysis

# Calculate the total quantity sold for each product.

# Which product appears to be the most popular?

# Ans :

df = pd.read_csv('/Advanced_pandas_sales_dataset.csv')

product_quantity = df.groupby('Product')['Quantity'].sum()
print(product_quantity)

popular_product = product_quantity.idxmax()
print("\nPopular Product : ",popular_product)

```

```python
# Q6 Multi-Level Grouping

# - Group the dataset by: Region and Category

# - Calculate the total sales for each combination.

# What insights can you derive from this analysis?

# Ans :

df = pd.read_csv('/Advanced_pandas_sales_dataset.csv')

grouped_data_sales = df.groupby(['Region', 'Category'])['Sales'].sum()
print(grouped_data_sales)
```

**Ans 6 :** **Insights**
- Electronics category contributes the highest sales in almost every region.
- East region performs best across categories.

```python
# Q7 Pivot Table Creation

# Create a pivot table showing:

# Rows → Region

# Columns → Product

# Values → Total Sales

# Which region sells the most laptops?


# Ans :

df = pd.read_csv('/Advanced_pandas_sales_dataset.csv')

pivot_table = pd.pivot_table(
    df,
    values='Sales',
    index='Region',
    columns='Product',
    aggfunc='sum'
)

print(pivot_table)


print("\nRegion that sells most laptop is : ", pivot_table['Laptop'].idxmax())

```

```python
# Q8 Profitability Analysis

# Calculate the average profit for each product category.

# Which category is most profitable?

# Ans :

df = pd.read_csv('/Advanced_pandas_sales_dataset.csv')

profit_by_category = df.groupby(['Product','Category'])['Profit'].mean()
print(profit_by_category)

most_profitable_category = profit_by_category.idxmax()
print("\nMost Profitable Category : ",most_profitable_category)


```

```python
# Q9 Discount Impact Study

# Analyze the relationship between discount and profit.

# Tasks:

# Calculate the average profit for each discount level

# Identify whether higher discounts lead to lower profit

#Ans :

df= pd.read_csv('/Advanced_pandas_sales_dataset.csv')

profit_by_discount = df.groupby('Discount')['Profit'].mean()
print(profit_by_discount)

```

- As discount increases :  average profit decreases
- Lower discounts (0–5%) : higher profit
- Higher discounts (15–20%) : lower profit

```python
# Q10 End-to-End Data Wrangling Task

# Perform the following workflow:

# 1. Filter the dataset where Sales > 3000

# 2. Group data by Region and Product

# 3. Calculate total sales

# 4. Reshape the result into a pivot table

# Explain the business insights obtained from this analysis.


# Ans :

#dataset with sales > 3000

df = pd.read_csv('/Advanced_pandas_sales_dataset.csv')

high_sales = df[df['Sales']>3000]
print("\nHigh Sales", "\n",high_sales)


# group by region and product

product_by_region = high_sales.groupby(['Region', 'Product'])['Sales'].sum()
print(product_by_region)


# Total Sales

total_sales = product_by_region.sum()
print("\nTotal Sales : ",total_sales)


# Pivot table

pivot_table = pd.pivot_table(
    high_sales,
    values='Sales',
    index='Region',
    columns='Product',
    aggfunc='sum'
)

print("\nPivot Table : ","\n",pivot_table
)
```
